### Beautifulsoup 

```python
html_doc = """
<html>
    <head>
        <title>Example Page</title>
    </head>
    <body>
        <h1>Welcome to Web Scraping</h1>
        <p class="description">This is an example paragraph.</p>
        <a href="https://example.com1">Visit Example1</a>
        <a href="https://example.com2">Visit Example2</a>
        <a href="https://example.com3">Visit Example3</a>
        <a href="https://example.com4">Visit Example4</a>
    </body>
</html>
"""
```

```python
# BeautifulSoup 객체를 생성 
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
print(soup.title.string)
print(soup.h1)
print(soup.a['href'])
print(soup.p['class'])
```

```python
# find(tag) , find_all(tag)
a_tag = soup.find('a')
print(a_tag)
a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []
for a_tag in a_tags:
    print(a_tag['href'])
    links.append(a_tag['href'])

print(links)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = webdriver.Chrome()

driver.get('https://www.google.com')

# 검색창 찾기
search_box = driver.find_element(By.NAME, 'q')

#검색어 처리
search_box.send_keys('python')
# 엔터
search_box.send_keys(Keys.RETURN)

# 결과가 나올때까지 멈춤
time.sleep(10)

results = driver.find_elements(By.XPATH, '//*[@id="rso"]/div[2]/div[5]/div/div/div/div/div[1]/div/div/span/a/h3')

print(results)
```

```python
driver = webdriver.Chrome()

url = 'https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=105'

driver.get(url)


time.sleep(3)

results = driver.find_elements(By.CSS_SELECTOR, 'ul.type06_headline li dt:not(.photo) a')
print(results)

for item in results:
    print(item.text)
    print(item.get_attribute('href'))

driver.quit()
```

### SQLite: 관계형데이터베이스

```python
import sqlite3
```

```python
con = sqlite3.connect('first.db')
```

```python
# 관계형데이터베이스: 테이블 구조로 데이터를 저장
# 저장할 수 있는 공간 생성 => table 생성
# sql : create table <테이블 명>(컬럼이름 데이터타입 제약조건, 컬럼)

# 테이블 생성
# 1. 데이터베이스 연결

sql_create= '''
create table user (
    id integer primary key autoincrement,
    name text not null,
    age integer,
    city text
)
'''

con = sqlite3.connect('first.db')

# 2. sql 실행할 수 있는 객체 필요
cursor = con.cursor()

cursor.execute(sql_create)

con.close()
```

```python
# 데이터 입력: 행단위 입력 => (이름, 나이, 거주지)
# insert into user (name, age, city) values(?,?,?)

# 1. 데이터베이스 연결
con = sqlite3.connect('first.db')

# 2. 데이터처리
cursor = con.cursor()

# sql문 작성
sql_insert = 'insert into user (name, age, city) values(?,?,?)'

# cursor.execute(sql_insert, ('cool', 20, 'London'))
data = [
    ('cool', 20, 'London'),
    ('hot',  30, 'Seoul'),
    ('king', 40, 'London'),
]

cursor.executemany(sql_insert, data)
# 입력, 수정, 삭제 -> DML
# 완료 여부 코드로 지정 => 물리지ㅓㄱ인 저장
con.commit()

con.close()
```

```python
# 질의 (SELECT): 데이터 검색 => 행단위 검색
# sql = SELECT 컬럼 ..fromm 테이블 이름 [where 컬럼이름]기준으로 비교 연
data = [
    ('cool', 20, 'London'),
    ('hot',  30, 'Seoul'),
    ('king', 40, 'London'),
]
#1.연결
con = sqlite3.connect('first.db')

cursor = con.cursor()
sql_select = 'select * from user'

cursor.execute(sql_select, data)

# slect 결과 =>fetchall() , fetchone()
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# 데이터 수정 행 단위 데이터 수정, 행을 찾고 -> 행의 컬럼의 값을 변겅(삭제, 수정,추가)


# 1. 연결
con = sqlite3.connect('first.db')

# 2. 데이터 처리
cursor = con.cursor()
sql_update = 'update user set age=? where id=?'

cursor.execute(sql_update, (60,1))

con.commit()
print('update ok')
```

```python
# 데이터의 삭제

con = sqlite3.connect('first.db')
cursor = con.cursor()
sql_delete = 'delete from user where id=?'
cursor.execute(sql_delete, (1,))
con.commit()

print('delete ok')
```
